# ML Final Project

```python
import numpy as np
from numpy import genfromtxt

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

### Some functions

```python
def RMSE(predict, y):
    err = predict - y
    sq_err = np.square(err)
    
    sum_sq_err = np.sum(sq_err, 0)
    sum_sq_err = sum_sq_err / np.shape(y)[0]
    sum_sq_err = np.sqrt(sum_sq_err)
    
    return sum_sq_err
```

```python
full_training = genfromtxt('full_data.csv', delimiter=',', dtype=np.dtype(float))

#First column are the y values
full_y = np.array(np.zeros((np.shape(full_training)[0], 1)))
for i in range(0, np.shape(full_training)[0]):
    full_y[i][0] = full_training[i][0]
    
#Rest are features
full_x = np.array(np.zeros((np.shape(full_training)[0], np.shape(full_training)[1])))
for i in range(0,np.shape(full_training)[0]):
    for j in range(1, np.shape(full_training)[1]):
        full_x[i][j-1] = full_training[i][j]
    full_x[i][np.shape(full_training)[1] - 1] = 1
    
print np.shape(full_x)
print np.shape(full_y)

training_x = full_x[:1000]
training_y = full_y[:1000]

test_x = full_x[1000:1500]
test_y = full_y[1000:1500]
```

## Ordinary Least Squares

```python
regr = linear_model.LinearRegression()
regr.fit(training_x, training_y)

predict = regr.predict(test_x)
#print predict - test_y

print RMSE(predict, test_y)
```

## Ridge Regression

```python
regr = linear_model.RidgeCV(alphas=[0.1, 1.0, 10.0])
regr.fit(training_x, training_y)

predict = regr.predict(test_x)

print RMSE(predict, test_y)
```

## Lasso

```python
regr = linear_model.Lasso(alpha = 0.1)
regr.fit(training_x, training_y)

predict = regr.predict(test_x)
#print predict - test_y

print RMSE(predict, test_y)
```
